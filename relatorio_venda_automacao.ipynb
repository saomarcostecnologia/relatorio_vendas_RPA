```python
#1° Entrar no sistema da empresa (Acessar link https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga)
#Comandos pyautogui: py.click, py.write, py.press, py.hotkey

import pyautogui as py
import pyperclip as pclip
import time
import pandas as pd

py.PAUSE = 1
py.FAILSAFE = True 

py.hotkey("Win")
py.write("Opera")
py.press("Enter")
pclip.copy("https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga")
py.hotkey("ctrl","v")
py.press("Enter")

time.sleep(3)

#2° Navegar até local do relatório (Entrar na pasta exportar)

py.click(x=387, y=348, clicks = 2) #Exportar
time.sleep(2)

#3° Exportar relatório (Vendas)
py.click(x=1263, y=344) #3 pontos
time.sleep(2)
py.click(x=1012, y=584) #Download
time.sleep(2)
py.click(x=788, y=482) #Confirmar download

#4° Calcular indicadores(Faturamento,qt de produtos vendidos, média de vendas por tipo de produtos)

tabela = pd.read_excel(r"C:\Users\glmarcos\Downloads\Vendas - Dez.xlsx")
#display(tabela) print de display da tabela indicada

    #4.1 Calculo dos indicadores
faturamento = tabela["Valor Final"].sum() #Soma coluna faturamento
quantidade = tabela["Quantidade"].sum() #Soma coluna quantidade

#print(faturamento) Validando resultado
#print(quantidade) Validando resultado
time.sleep(2)

#5° Enviar email (gabriel.433@hotmail.com)
    #5.1 Acessar gmail
py.hotkey("ctrl","t")
pclip.copy("https://mail.google.com/mail/u/0/#inbox")
py.hotkey("ctrl","v")
py.hotkey("Enter")
time.sleep(1)

    #5.2 Escrever email
py.click(x=106, y=185)
py.write("gabriel.433@hotmail.com")
py.hotkey("Enter")
time.sleep(1)

    #5.3 Escrever assunto
py.hotkey("Tab")
pclip.copy("Relatórios de Vendas")
py.hotkey("ctrl","v")
py.hotkey("Tab")
time.sleep(1)

    #5.4 Escrever corpo email
texto = f"""Prezados,

Segue relatório de Vendas: 

Faturamento : R$ {faturamento:,.2f}
Quantidade : {quantidade:,}

Qualquer dúvida fico à disposição.

Att., 

Gabriel São Marcos
"""
    #5.5 Enviar email
pclip.copy(texto)
py.hotkey("ctrl","v")

time.sleep(1)
py.hotkey("tab")
py.hotkey("enter")

#Para rodar todo dia tal horario? #Executar codigo automaticamente (2 videos de treino sobre deixar rodando auto e com pc desligado)
#Para transformar o arquivo em executavel ? ver video de como trasnformar arquivo python em executavel

```
